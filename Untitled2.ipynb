```python
df = df.drop('default', axis = 1)
```

```python
def age_split(df) :
    if df['age'] <= 20 :
        return '10대'
    elif (df['age'] > 20) & (df['age'] <= 30) :
        return '20대'
    elif (df['age'] > 30) & (df['age'] <= 40) :
        return '30대'
    elif (df['age'] > 40) & (df['age'] <= 50) :
        return '40대'
    elif (df['age'] > 50) & (df['age'] <= 60) :
        return '50대'
    elif (df['age'] > 60) & (df['age'] <= 70) :
        return '60대'
    elif (df['age'] > 70) & (df['age'] <= 80) :
        return '70대'
    elif (df['age'] > 80) & (df['age'] <= 90) :
        return '80대'
    elif (df['age'] > 90) & (df['age'] <= 100) :
        return '90대'
df['age_group'] = df.apply(age_split, axis = 1)
```

```python
target_col = ['deposit']
```

```python
bool_columns = ['housing', 'loan']
cat_columns = ['job', 'marital', 'education', 'contact', 'month', 'poutcome','age_group']
```

```python
num_cols = [x for x in df.columns if x not in target_col + bool_columns + cat_columns]
num_cols
```

```python
df['previous'].value_counts()
```

```python
df[num_cols]
```

```python
# bool ohe
columns = ['housing', 'loan','job', 'marital', 'education', 'contact', 'month', 'poutcome']

for col in columns:
    df = df.merge(pd.get_dummies(df[col], prefix=col),
             left_on=df.index, right_on=pd.get_dummies(df[col], 
                                                       prefix=col).index)
    df = df.drop('key_0', axis=1)
df
```

```python
df = df.drop(columns, axis=1)
```

```python
df['deposit'] = df['deposit'].replace('yes',1).replace('no', 0)
```

```python
X = df.drop(columns = 'deposit')
y = df[['deposit']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,
                                                   random_state=42)
```

```python
X_train
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(random_state=42)
model
```

```python
model.fit(X_train, y_train)
predictions = model.predict(X_test)
```

```python
predictions[:5]
```

```python
from sklearn.metrics import f1_score

f1_score(y_test, predictions) * 100
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, predictions) * 100
```
